```python
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd 
import os 
```

```python
solar_main_path = os.path.join(os.getcwd(), "Mean_GHI_Shp.shp")
solar_sample_path = os.path.join(os.getcwd(), "Mean_GHI_Random_Sample.shp")
counties_path = os.path.join(os.getcwd(), "US_counties.shp")
print(os.path.exists(solar_main_path), os.path.exists(solar_sample_path), os.path.exists(counties_path))
```

```python
solar_main = gpd.read_file(solar_main_path)
solar_main = solar_main.to_crs("EPSG:4326")
solar_main['geometry'] = solar_main.geometry.round(6)
solar_main.head(10)
```

```python
counties = gpd.read_file(counties_path)
counties = counties.to_crs("EPSG:4326")
counties['geometry'] = counties.geometry.round(6)
counties.head(10)
```

```python
solar_main_join = solar_main.sjoin(counties, predicate="intersects")
solar_main_join.head(10)
```

{'overlaps', 'touches', 'contains_properly', 'intersects', 'within', None, 'covered_by', 'contains', 'covers', 'crosses'}

```python
# solar_main_join = solar_main_join.drop_duplicates(subset="FIPS", keep="first")
solar_main_join = solar_main_join.dissolve(by="FIPS")
```

```python
solar_main_join.head(10)
```

```python
arizona = solar_main_join[solar_main_join['STATE'] == "AZ"]
arizona.head(20)
```
