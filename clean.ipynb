```python
income_path = r"US Median Household Income by County.csv" 
degree_path = r"US Population with Bachelor's Degree by Bachelors.csv"
print(os.path.exists(income_path), os.path.exists(degree_path))
```

## Income Cleaning

```python
income_df = pd.read_csv(income_path, skiprows=4)
income_df.head(10)
```

Pop the `United States` row.

```python
income_co_df = income_df.drop(index=0)
income_co_df.head(10)
```

Split the `County` column into `County_name` and `State`.

```python
test_string = "stupid, boy"
print(test_string.split(", ")[0])
print(test_string.split(", ")[1])
print(len(test_string.split(", ")))
```

```python
for i, row in income_co_df.iterrows():
  parts = row['County'].split(", ")
  if len(parts) >1:
    income_co_df.at[i, 'County_Name'] = parts[0]
    income_co_df.at[i, "State"] = parts[1]
  else:
    income_co_df.at[i, "State"] = parts[0]
```

```python
income_co_df.head(10)
```

```python
income_co_df = income_co_df.drop(income_co_df.tail(8).index)
```

```python
income_co_df.tail(10)
```

```python
income_co_df['FIPS'] = income_co_df['FIPS'].astype(np.int32)
income_co_df.head(10)
```

## US Counties Shapefile.

```python
import geopandas as gpd
```

```python
counties_path = r"US Counties.shp"
print(os.path.exists(counties_path))
```

```python
counties_df = gpd.read_file(counties_path)
counties_df.head(10)
```

```python
counties_df['FIPS'] = counties_df['FIPS'].astype(np.int32)
counties_df.head(10)
```

```python
income_counties_merge = counties_df.merge(income_co_df, on="FIPS")
income_counties_merge.head(10)
```

```python
degree_counties_merge = counties_df.merge(degree_co_df, on="FIPS")
degree_counties_merge.head(10)
```

```python
income_counties_merge.to_file("US Median Household Income by County Cleaned.shp")
degree_counties_merge.to_file("US Population with Bachelor's Degree by County Cleaned.shp")
```

```python
income_counties_merge_df = pd.DataFrame(income_counties_merge.drop(columns='geometry'))
degree_counties_merge_df = pd.DataFrame(degree_counties_merge.drop(columns='geometry'))
```

## Solar Radiation

```python
import rasterio 
```

```python
ghi_path = r"C:\Users\96653\OneDrive - University of Arizona\Documents\DATA 363\Education vs sunny weather\Rasters\Mean_GHI_County.tif"
print(os.path.exists(ghi_path))
```

```python
with rasterio.open(ghi_path) as src:
  solar_data = src.read(1)
```

```python
counties_df = counties_df.to_crs(src.crs)
```

```python
from rasterstats import zonal_stats
```

```python
stats = zonal_stats(counties_df, solar_data, affine=src.transform, stats=['mean'])
counties_df['GHI'] = [s['mean'] for s in stats]
```

```python
ghi_result_df = pd.DataFrame({
    'County': counties_df['COUNTYNAME'],
    'FIPS': counties_df['FIPS'],
    'GHI': counties_df['GHI']
})
```

```python
ghi_result_df.to_csv('solar_radiation_by_county.csv', index=False)
```
