## T-test

```python
def t_stat(sample_mean, population_mean, sample_std, sample_size):
  return (sample_mean-population_mean)/(sample_std/np.sqrt(sample_size))
```

```python
population_mean = np.mean(arizona_main['Value__Per'])

sample_50_mean = np.mean(sample_50['Value__Per'])
sample_50_std = np.std(sample_50['Value__Per'])

sample_100_mean = np.mean(sample_100['Value__Per'])
sample_100_std = np.std(sample_100['Value__Per'])

sample_500_mean = np.mean(sample_500['Value__Per'])
sample_500_std = np.std(sample_500['Value__Per'])

sample_5000_mean = np.mean(sample_5000['Value__Per'])
sample_5000_std = np.std(sample_5000['Value__Per'])
```

```python
print(t_stat(sample_50_mean, population_mean, sample_50_std, 50))
print(t_stat(sample_100_mean, population_mean, sample_100_std, 100))
print(t_stat(sample_500_mean, population_mean, sample_500_std, 500))
print(t_stat(sample_5000_mean, population_mean, sample_5000_std, 5000))
```
